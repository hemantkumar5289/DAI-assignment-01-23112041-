```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('netflix_titles.csv')
data
```

```python
data.isnull().sum()
```

```python
data['director'] = data['director'].fillna('No Data Availabe')
data['cast'] = data['cast'].fillna('No Data Available')
data['country'] = data['country'].fillna('No Data Available')
data['date_added'] = data['date_added'].fillna(data['date_added'].mode()[0])  
data['rating'] = data['rating'].fillna(data['rating'].mode()[0])
data['duration'] = data['duration'].fillna('No Data Available')
```

```python
TV_shows =data.loc[data['type'] == "TV Show"].count()[0]
movies = data.loc[data['type'] == 'Movie'].count()[0]                                             
colors=('red','blue')
labels =['TV Show','Movie']
plt.title("Netflix Content in %",fontdict = {'fontweight':'bold','fontsize':18})
piechart = plt.pie([TV_shows,movies],colors=colors,labels=labels,autopct ='%.2f %%')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
data.country.value_counts()[:10].plot(kind="bar")
plt.title('Top 10 countries with Most Contributions to Netflix',fontdict={"fontweight":"bold"})
plt.xlabel('Country')
plt.ylabel('')
plt.show()
```

```python
data.director.value_counts()[1:11].plot(kind="bar")
plt.yticks(np.arange(0,22,1))
plt.title("Top 10 directors with Most Netflix Shows",fontdict={'fontweight':'bold'})
plt.xlabel("Directors")
plt.ylabel("Total Shows")
plt.show()
```

```python
x = list(data.rating.value_counts().index)
y=list(data.rating.value_counts().values)
plt.subplots(figsize =(16, 9))
plt.bar(x,y,color ='maroon')
#displaying values on top of the bar
for i, v in enumerate(y):
    plt.text(i,v+60,str(v),ha ="center",weight="bold",size=10)
plt.title('Distribution of Ratings')
plt.show()
```
